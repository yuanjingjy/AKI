```python
from sklearn.feature_selection import f_regression, mutual_info_regression
import numpy as np
f_test, _ = f_regression(datamat, labelmat)
f_test /= np.max(f_test)

mi = mutual_info_regression(datamat, labelmat)
mi /= np.max(mi)

plt.plot(f_test, label='f test')
plt.plot(mi,label='mi')
plt.legend(loc='upper right')
plt.show()
```

```python
from sklearn.feature_selection import SelectKBest, f_classif,mutual_info_classif
selector_f = SelectKBest(f_classif,k=10)
selector_f.fit(datamat, labelmat)
score_f = selector_f.pvalues_
# score_f /= score_f.max()

selector_mi = SelectKBest(mutual_info_classif,k=10)
selector_mi.fit(datamat, labelmat)
score_mi = selector_mi.pvalues_
# score_mi /= score_mi.max()

# plt.plot(score_f, label='f test')
# plt.plot(score_mi,label='mi')
# plt.legend(loc='upper right')
# plt.show()
```
